```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
market_data_scaled = StandardScaler().fit_transform(df_market_data[['price_change_percentage_24h', 'price_change_percentage_7d',
                                                                    'price_change_percentage_14d', 'price_change_percentage_30d',
                                                                    'price_change_percentage_60d', 'price_change_percentage_200d',
                                                                    'price_change_percentage_1y']])
```

```python
# Create a DataFrame with the scaled data
market_data_scaled_df = pd.DataFrame(market_data_scaled, columns=['price_change_percentage_24h', 'price_change_percentage_7d',
                                                                    'price_change_percentage_14d', 'price_change_percentage_30d',
                                                                    'price_change_percentage_60d', 'price_change_percentage_200d',
                                                                    'price_change_percentage_1y'])

# Copy the crypto names from the original data
market_data_scaled_df['coin_id'] = df_market_data.index

# Set the coinid column as index
market_data_scaled_df = market_data_scaled_df.set_index('coin_id')

# Display sample data
market_data_scaled_df.head()
```

---

### Find the Best Value for k Using the Original Data.

```python
# Create a list with the number of k-values from 1 to 11
k = list(range(1, 11))
```

```python
# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_scaled`
# 3. Append the model.inertia_ to the inertia list

for value in k:
    k_model = KMeans(n_clusters=value, n_init='auto', random_state=1)
    k_model.fit(market_data_scaled_df)
    inertia.append(k_model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
orig_elbow_dict = {
    'K': k,
    'inertia': inertia
}

# Create a DataFrame with the data to plot the Elbow curve

orig_elbow_df = pd.DataFrame(orig_elbow_dict)
orig_elbow_df
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
hvplot.extension('bokeh')
orig_elbow_plot = orig_elbow_df.hvplot.line(
    x="K",
    y="inertia",
    xticks=k,
    title = "Original Data"
)
orig_elbow_plot
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** The optimal value for k is 4. This is the point at which the "elbow" in the curve appears, where the data points become tightly clustered.

---

### Cluster Cryptocurrencies with K-means Using the Original Data

```python
# Initialize the K-Means model using the best value for k
orig_model = KMeans(n_clusters=4, n_init='auto', random_state=1)
```

```python
# Fit the K-Means model using the scaled data
orig_model.fit(market_data_scaled_df)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled data
scaled_clusters_predicted = orig_model.predict(market_data_scaled_df)

# Print the resulting array of cluster values.
scaled_clusters_predicted
```

```python
# Create a copy of the DataFrame
scaled_clusters_predicted_df = market_data_scaled_df.copy()
```

```python
# Add a new column to the DataFrame with the predicted clusters
scaled_clusters_predicted_df["group"] = scaled_clusters_predicted

# Display sample data
scaled_clusters_predicted_df.head()
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.

hvplot.extension('bokeh')
scaled_clusters_plot = scaled_clusters_predicted_df.hvplot.scatter(
    x="price_change_percentage_24h",
    y="price_change_percentage_7d",
    by="group",
    hover_cols=['coin_id'],
    title = "Crypto Clusters - Scaled Data"
)
scaled_clusters_plot

# To assure that the crypto name was correctly added to the hover_cols parameter, the following source was consulted: 
# https://stackoverflow.com/questions/59678780/show-extra-columns-when-hovering-in-a-scatter-plot-with-hvplot
```

---

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca = PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` to reduce to 
# three principal components.
pca_clusters_predicted = pca.fit_transform(scaled_clusters_predicted_df)

# View the first five rows of the DataFrame. 
pca_clusters_predicted[:5]
```

```python
# Retrieve the explained variance to determine how much information 
# can be attributed to each principal component.
pca.explained_variance_ratio_
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** The total explained variance is .88862186, or approximately 88.9%. And of that total variance, approximately 34.9% is explained by the 
first principal component, 31.4% by the second component, and 22.6% by the third.

```python
# Create a new DataFrame with the PCA data.

# Creating a DataFrame with the PCA data
pca_clusters_predicted_df = pd.DataFrame(pca_clusters_predicted, columns=['PC1', 'PC2', 'PC3'])

# Copy the crypto names from the original data
pca_clusters_predicted_df ['coin_id'] = df_market_data.index

# Set the coinid column as index
pca_clusters_predicted_df = pca_clusters_predicted_df.set_index('coin_id')

# Display sample data
pca_clusters_predicted_df.head()
```

---

### Find the Best Value for k Using the PCA Data

```python
# Create a list with the number of k-values from 1 to 11
k = (range(1, 11))
```

```python
# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_pca`
# 3. Append the model.inertia_ to the inertia list

for value in k:
  pca_model = KMeans(n_clusters = value, n_init='auto', random_state=1)
  pca_model.fit(pca_clusters_predicted_df)
  inertia.append(pca_model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
pca_elbow_dict = {
    'K': k,
    'inertia': inertia
}

# Create a DataFrame with the data to plot the Elbow curve
pca_elbow_df = pd.DataFrame(pca_elbow_dict)
pca_elbow_df.head()
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
hvplot.extension('bokeh')
pca_elbow_plot = pca_elbow_df.hvplot.line(
    x='K',
    y='inertia',
    xticks=k,
    title = "PCA Data"
)
pca_elbow_plot
```

#### Answer the following questions: 

* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:** The best value for k, using the PCA data, is again, 4.


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:** The k-value using PCA data does not differ from the k-value using the original data. Both show 4 as being the optimal value.

### Cluster Cryptocurrencies with K-means Using the PCA Data

```python
# Initialize the K-Means model using the best value for k
best_k_model = KMeans(n_clusters=4, n_init='auto', random_state=1)
```

```python
# Fit the K-Means model using the PCA data
best_k_model.fit(pca_clusters_predicted_df)
```

```python
# Predict the clusters to group the cryptocurrencies using the PCA data
clusters_group_crypto = best_k_model.predict(pca_clusters_predicted_df)

# Print the resulting array of cluster values.
clusters_group_crypto
```

```python
# Create a copy of the DataFrame with the PCA data
clusters_group_crypto_df = pca_clusters_predicted_df.copy()

# Add a new column to the DataFrame with the predicted clusters
clusters_group_crypto_df["group"] = clusters_group_crypto

# Display sample data
clusters_group_crypto_df.head()
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="PC1"` and `y="PC2"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.

hvplot.extension('bokeh')
crypto_clusters_plot = clusters_group_crypto_df.hvplot.scatter(
    x='PC1',
    y='PC2',
    by='group',
    hover_cols=['coin_id'],
    title = "Crypto Clusters - PCA Data"
)
crypto_clusters_plot
```
